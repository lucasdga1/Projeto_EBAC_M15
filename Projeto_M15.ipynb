```python
# Relação entre Gênero e tempo como cliente
gen_cli = df.groupby('Genero')['Tempo_como_Cliente'].mean().reset_index()
fig = px.bar(gen_cli,
             y = 'Tempo_como_Cliente',
             x = 'Genero',
             color = 'Genero')
fig.show()
```

Nesse gráfico é possível concluir que o tempo como cliente está equilibrado entre os gêneros e sua média é de 32 meses.

```python
si_tc = df.groupby('Servico_Internet')['Tipo_Contrato'].value_counts(normalize = True).rename('Proporcao').reset_index()
fig = px.bar(si_tc, x = 'Servico_Internet',
             y = 'Proporcao',
             facet_col = 'Tipo_Contrato',
             color = 'Servico_Internet')
fig.show()
```

 A fibra ótica domina os contratos mensais, mas os contratos longos demonstram uma preferência por não obter o serviço

```python
idoso_churn = df.groupby('Idoso')['Churn'].value_counts(normalize = True).rename('Proporcao').reset_index()
fig = px.histogram(idoso_churn,
             x = 'Churn',
             y = 'Proporcao',
             color = 'Churn',
             facet_col='Idoso')
fig.show()
```

Uma proporção maior de idosos tendem a dar churn, do que não idosos. Porém a proporção dos que ficam, mesmo sendo idosos, é maior.

```python
tc_churn = df.groupby('Tipo_Contrato')['Churn'].value_counts().reset_index()
fig = px.bar(tc_churn,
          x = 'Tipo_Contrato',
          y = 'count',
          barmode = 'stack',
          color = 'Churn')
fig.show()
```

Existe uma tendência maior a churn em contratos mensais

```python
pag_churn = df.groupby('Metodo_Pagamento')['Churn'].value_counts(normalize=True).reset_index()
fig = px.bar(pag_churn,
             x = 'Metodo_Pagamento',
             y = 'proportion',
             color = 'Churn')
fig.show()
```

O método com maior incidência de churn é o eletronic check, enquanto os outros se mantém balanceados

As variáveis relacionadas à idade do cliente, método de pagamento e duração do contrato demonstraram ter maior relevância para a incidência de churn.
